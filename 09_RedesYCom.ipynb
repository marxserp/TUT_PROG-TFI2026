```python
# Importar librerías Networkx (nodos y bordes) y Matplotlib (gráficas)
import networkx as nx
import matplotlib.pyplot as mpl
# Declarar grafo como nuevo Grafo
grafo = nx.Graph()
# Añadir 8 nodos (add_nodes) y establecer las relaciones (add_edges)
grafo.add_nodes_from(["Nodo1", "Nodo2", "Nodo3", "Nodo4", "Nodo5", "Nodo6", "Nodo7", "Nodo8"])
grafo.add_edges_from([("Nodo1","Nodo2"), ("Nodo2","Nodo3"), ("Nodo3","Nodo4"), ("Nodo4","Nodo5"), ("Nodo5","Nodo6"), ("Nodo6","Nodo7"), ("Nodo7","Nodo8"), ("Nodo8","Nodo1")])
# Demostración de cantidad de nodos y relaciones
print(f"Cantidad total de Nodos: {grafo.number_of_nodes()}")
print(f"Cantidad de relaciones: {grafo.number_of_edges()}")
# Definir tamaño y otros parámetros de la gráfica
mpl.figure(figsize=(5, 3))
nx.draw(grafo,
        with_labels=True,
        node_color='#5c6282',
        node_size=1500,
        font_weight='semibold',
        font_size=10,
        font_color="#e6eaff")
# Mostrar la gráfica
mpl.show()
```
